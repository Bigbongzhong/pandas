```python
import matplotlib.pyplot as plt 

x=[1,2,3,4,5]
y=[1,4,9,16,25]
labels =[0,1,0,1,0]

plt.scatter(x,y,c=labels)
plt.plot(x, y, 'rp')  # 'ro' means red color with circle markers
plt.title("scatter plotting")
plt.xlabel("numbers")
plt.ylabel("squares")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

fig, axs = plt.subplots(1, 2)  # 1 row, 2 columns
fig.suptitle("sin and cos waves")
axs[0].plot(x, y1)
axs[0].set_title("Sine Wave")

axs[1].plot(x, y2)
axs[1].set_title("Cosine Wave")

plt.show()

```
