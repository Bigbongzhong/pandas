```python
import pandas as pd

df = pd.read_csv(r"C:\Users\ohohu\Downloads\train.csv.csv")
df.head()
```

```python
df.drop(['PassengerId', 'Name', 'SibSp', 'Parch','Ticket','Cabin','Embarked'], axis='columns', inplace=True)
df.head()
```

```python
#made independent and dependent variables named as inputs and target respectively
target = df.Survived
inputs = df.drop('Survived', axis='columns')
```

```python
# Convert Sex comlumn to neumerical values like 0 and 1
dummies = pd.get_dummies(inputs.Sex).astype(int)
dummies.head(3)
```

```python
print(inputs.to_string())
```

```python
inputs = inputs.loc[:, ~inputs.columns.duplicated()]
```

```python
print(inputs.to_string())
```

```python
inputs.columns[inputs.isna().any()]
```

```python
inputs.Age = inputs.Age.fillna(inputs.Age.mean())
inputs.head(6)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs, target, test_size=0.2)
```

```python
len(X_train), len(X_test), len(y_train), len(y_test)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train, y_train)
```

```python
y_test[:10]
```

```python
model.score(X_test, y_test)
```
